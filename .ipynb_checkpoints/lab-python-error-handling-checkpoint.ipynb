def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                valid_quantity = True
            except ValueError as error:
                print(f"Error: {error}")
        inventory[product] = quantity
    return inventory

````python
products = ["t-shirt", "mug", "hat", "book", "keychain"]

def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                valid_quantity = True
            except ValueError as error:
                print(f"Error: {error}")
        inventory[product] = quantity
    return inventory

#2 Modify the calculate_total_price function to include error handling.

def total_price(customer_orders, inventory):
    if not customer_orders:
        print('No products ordered.')
        return 0

    valid_orders = {product: quantity for product, quantity in customer_orders.items() if product in inventory and inventory[product] > 0}

    if not valid_orders:
        print("None of the ordered products are in stock.")
        return 0

    total = 0
    prices = {}
    for product, quantity in valid_orders.items():
        while True:
            try:
                price = float(input(f"Please enter the price for {product}: "))
                if price < 0:
                    raise ValueError("Price cannot be negative.")
                prices[product] = price 
                break
            except ValueError:
                print(f"Invalid input for {product}. Please enter a valid numerical value.")

        total += prices[product] * quantity

    print(f"Total price of the order: {total}")
    print(f"Updated Inventory:")
    for product, quantity in inventory.items():
        print(f"{product}: {quantity}")
    
    return total
#3 Modify the get_customer_orders function to include error handling.

def customer_orders(inventory):
    correct = False
    while not correct:
        try:
            num_orders = int(input("How many different products would you like to order? "))
            if num_orders <= 0:
                raise ValueError("The number of products must be greater than zero.")

            customer_order = {}

            for i in range(num_orders):
                while True:
                    product_name = input(f"Please enter the name of product #{i+1}: ").lower()
                        
                    if product_name not in inventory.keys():
                        print("Sorry, this product is not in the inventory. Please refer to the list below:")
                        print(list(inventory.keys()))
                    else:
                        if inventory[product_name] >= 1:
                            customer_order[product_name] = 1
                            inventory[product_name] -= 1
                            break
                        else:
                            print(f"Sorry, {product_name} is out of stock. Please choose another product.")
            
            print("Order placed:")
            for product_name, quantity in customer_order.items():
                print(f"{product_name}: {quantity}")
            correct = True
        except ValueError:
            print("Invalid input. Please enter a numerical value.")
    return customer_order

#4 Test your code by running the program and deliberately entering invalid quantities and product names. 
#Make sure the error handling mechanism works as expected.

inventory = initialize_inventory(products)
customer_orders = customer_orders(inventory)
total = total_price(customer_orders, inventory)
````
